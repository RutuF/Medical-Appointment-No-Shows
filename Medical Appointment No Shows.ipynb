```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\Rutushikha Fulzele\Downloads\archive (5)\KaggleV2-May-2016.csv")
df
```

```python
#Data cleaning
```

```python
df = df.drop(columns = ["Neighbourhood"])
df
```

```python
df.dtypes
```

```python
print(df.isna().sum())
```

```python
#Clean Anomalies and Outliers

# Remove rows where Age < 0
df = df[df['Age'] >= 0]
df
```

```python
#"No-show" has values: "No" and "Yes" → convert to binary

df['No-show'] = df['No-show'].replace({'No': 0, 'Yes': 1})
df['No-show']
```

```python
df['Gender'] = df['Gender'].replace({'F': 0, 'M' : 1})
df['Gender']
```

```python
No_show_target = df['No-show']
No_show_target
```

```python
df_features = df.drop(['No-show'], axis = 1)
df_features
```

```python
df_features['ScheduledDay'] = pd.to_datetime(df_features['ScheduledDay'])
df_features['ScheduledDay']
```

```python
df_features['AppointmentDay'] = pd.to_datetime(df_features['AppointmentDay'])
df_features['AppointmentDay']
```

```python
df_features['WaitDays'] = (df_features['AppointmentDay'] - df_features['ScheduledDay']).dt.days
```

```python
#.dt.dayofweek extracts the day of the week as a number from the datetime.

df_features['ScheduledWeekday'] = df_features['ScheduledDay'].dt.dayofweek
df_features['AppointmentWeekday'] = df_features['AppointmentDay'].dt.dayofweek
```

```python
df_features = df_features.drop(['ScheduledDay', 'AppointmentDay'], axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="No-show", data = df)
plt.title("Show Vs No-show")
plt.xticks([0, 1], ['Show', 'No-show'])
plt.show()
```

```python
#by gender

sns.countplot(x= 'Gender', hue ='No-show', data = df)
plt.title("No-show by Gender")
plt.show()

#females are more likely to attend
```

```python
sns.histplot(data = df, x = 'Age', hue= 'No-show', bins = 15, kde = True, element = 'step')
plt.title("Age by No-show")
plt.show()
#no-shows are spread across many ages, but still less frequent.
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_features, No_show_target, test_size = 0.2, random_state = 42)
```

```python
#Logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

#Model is biased toward predicting class 0 (Show). Even though accuracy is 80%, it completely ignores the No-show class.

#This is likely because the data is imbalanced: only ~20% are No-shows.
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred, zero_division=0))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state = 42)
rf.fit(X_train, y_train)
```

```python
y_pred_rf = rf.predict(X_test)
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))

#Model predicts class 0 well (people who showed up), with: 96% recall, 82% precision. But it struggles with class 1 (no-shows):
#Only 15% recall → it misses 85% of actual no-shows
#48% precision → of those predicted no-shows, over half are wrong
```

```python
import xgboost as xgb
from sklearn.metrics import classification_report, confusion_matrix

# Calculate scale_pos_weight
scale = y_train.value_counts()[0] / y_train.value_counts()[1]

model = xgb.XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    scale_pos_weight=scale,  # Important for imbalance
    use_label_encoder=False,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

#Class 1 Recall Improved: XGBoost is capturing 75% of class 1 (no-shows) — which is a significant improvement from earlier models like Random Forest.

#Precision Trade-off: Class 1 precision is low (0.31), meaning many predictions of “no-show” are incorrect.
```

```python
#If the goal is not to miss any no-shows (i.e., high recall for class 1), this model performs reasonably well. 
#For example, in medical or appointment settings where catching no-shows is more critical than over-alerting, this model makes sense.
```
